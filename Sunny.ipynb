```python
import numpy as np
import pandas as pd
import datetime as dt
```

```python
df=pd.read_csv("crash_data.csv")
df=df.fillna(0)
df['fatalities']=df['fatalities'].astype(str).map(eval)
df['date'] = pd.to_datetime(df['date'], errors='coerce')
df["month"]=df["date"].dt.month
df["year"]=df["date"].dt.year
df.head()
```

```python
country_fatalities=df[["country","fatalities","year","month"]]
new_df=country_fatalities[(country_fatalities["year"]>=2000) & (country_fatalities["year"]<=2019)]
new_df.head()
```

```python
max_fac=new_df.groupby("country").agg({"fatalities":np.sum}).reset_index()
max_fac.sort_values(by="fatalities",ascending=False,inplace=True)
top_countries=max_fac[:10]
top_countries=pd.merge(new_df,top_countries["country"],on="country",how="inner")
top_countries
```

```python
pivot_monthyear=top_countries.pivot_table(index=["country"],values="fatalities",columns=["year","month"],aggfunc=np.sum).reset_index().fillna(0)
pivot_month=top_countries.pivot_table(index=["country","year"],values="fatalities",columns="month",aggfunc=np.sum).reset_index().fillna(0)
pivot_month
```
